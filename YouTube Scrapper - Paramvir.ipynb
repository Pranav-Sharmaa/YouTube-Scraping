```python
from bs4 import BeautifulSoup
import requests
from selenium import webdriver


import numpy as np
import pandas as pd
```

```python
browser=webdriver.Chrome()

browser.get('https://www.youtube.com/@PassengerParamvir/videos')
```

```python
soup=BeautifulSoup(browser.page_source,'html.parser')
```

```python
data=[]

for sp in soup.find_all("ytd-rich-item-renderer"):
    try:
        title = sp.find('a', class_='yt-simple-endpoint focus-on-expand style-scope ytd-rich-grid-media').text
    except AttributeError:
        title = None

    try:
        link = sp.find('a', class_='yt-simple-endpoint focus-on-expand style-scope ytd-rich-grid-media').get('href')
    except AttributeError:
        link = None

    try:
        views = sp.find('span', class_="inline-metadata-item style-scope ytd-video-meta-block").text
    except AttributeError:
        views = None

    try:
        thumbnail_link = sp.find('img').get('src').split('?')[0]
    except (AttributeError, IndexError):
        thumbnail_link = None

    data.append([title, views,link, thumbnail_link])
```
