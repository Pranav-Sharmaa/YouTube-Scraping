```python
import time
import numpy as np
import pandas as pd
from tqdm import tqdm

from bs4 import BeautifulSoup
from selenium import webdriver
```

```python
df = pd.read_csv('data.csv')

df.head()
```

```python
browser=webdriver.Chrome()
data=[]

for link in tqdm(df['link']):
    

    link='https://www.youtube.com/'+link

    browser.get(link)
    
    time.sleep(2)


    soup=BeautifulSoup(browser.page_source,'html.parser')
    
    
    
    try:
        
        views =soup.find("span",class_="bold style-scope yt-formatted-string").text
    except:
        views = np.nan

    
    try:
        title=soup.find_all('yt-formatted-string',class_='style-scope ytd-watch-metadata')[0].text
    except:
        title=np.nan
        
    try:
        
        likes =soup.find_all('div',class_='yt-spec-button-shape-next__button-text-content')[4].text
    except:
        likes=np.nan
        
        
    try:
        
        video_time=soup.find_all('span','bold style-scope yt-formatted-string')[2].text
    except:
        video_time=np.nan
        
    try:
        
        description=soup.find('span',class_='yt-core-attributed-string yt-core-attributed-string--white-space-pre-wrap').text
    
    except:
        description=np.nan
   
    
    data.append([title , video_time, likes, views, description])
```

```python
df = pd.DataFrame(data, columns = ['title', 'video_time', 'likes', 'views', 'description'])
```
